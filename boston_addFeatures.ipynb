# **Importing the libraries**

```python
import numpy as np
from sklearn import datasets,model_selection
import pandas as pd
from sklearn.linear_model import LinearRegression
```

# **data variable mei took boston dataset and analysed** 

```python
boston_data=datasets.load_boston()
boston_data
```

# **So all inputs in data and output in target aur column headers in feature_names**

```python
input=boston_data.data
output=boston_data.target
dataframe=pd.DataFrame(input)
#dataframe has all input data as pandas table/dataframe
#now change the column headers
dataframe.columns=boston_data.feature_names
dataframe
```

# **Created x_train,x_test,y_train,y_test**

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(input,output,random_state=0)
x_train #so output se clear that x_train and all are arrays
```

# **ek algorithm created to fit the x_train and y_train values**

```python
alg1=LinearRegression() #so alg1 is object of Linear Regression
alg1.fit(x_train,y_train)
print("All slopes/coefficient corresponding to different features : \n",alg1.coef_) #itne saare coefficient as all features ke respect mei ek optimal coefficient(m1,m2,m3,......,m13)
print("\nIntercept of all the fitting lines : \n",alg1.intercept_) #but intercept ek hi hoga saare fitting lines ka so all fitting lines have same intercept different slopes
```

# **Outputing the score**

```python
score_test=alg1.score(x_test,y_test)
score_train=alg1.score(x_train,y_train)
print("Training Score : ",score_train)
print("Testing Score : ",score_test)
```

```python
k=26
for i in range(len(boston_data.feature_names)):
  for j in range(i+1,len(boston_data.feature_names)):
    dataframe[k]=dataframe[boston_data.feature_names[i]]*dataframe[boston_data.feature_names[j]]
    k=k+1
dataframe #104 columns (ie. 13C2+13+13(13 of initial))
```

```python
algo2=LinearRegression()
new_input=dataframe
x_train,x_test,y_train,y_test=model_selection.train_test_split(new_input,output)
algo2.fit(x_train,y_train)
score_training=algo2.score(x_train,y_train)
score_testing=algo2.score(x_test,y_test)
print("Training score : ",score_training)
print("Testing score : ",score_testing)
#so improved a lot
```
